```python

import sys
import os
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '../../src')))

# Import necessary functions from train_classifier
from classification.train_classifiers import load_and_process_data, scale_data, apply_smote, evaluate_classifiers


# Define file paths and model parameters
filepath = '../../data/combined_features/Model_4_Radiomics_Deep_Clinical.csv'
model_name = 'Radiomics_Only'
output_file = '../../results/Model_4_Radiomics_Deep_Clinical_Results.csv'

# Load and process data
X_train, y_train, X_test, y_test = load_and_process_data(filepath)

# Scale the data
X_train_scaled, X_test_scaled = scale_data(X_train, X_test)

# Apply SMOTE for class balancing
X_train_oversampled, y_train_oversampled = apply_smote(X_train_scaled, y_train)

# Evaluate classifiers and save results
evaluate_classifiers(X_train_oversampled, y_train_oversampled, X_test_scaled, y_test, model_name, output_file)
```
